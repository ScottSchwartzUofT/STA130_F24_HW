```python
url = "https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv"
baby_names = pd.read_csv(url)
baby_names.info()  # This will show you if there are any missing data
```

```python
# Check the number of missing values per column
missing_counts = baby_names.isnull().sum()

# Check the proportion of missing values per column
missing_proportion = baby_names.isnull().mean()

print("Missing Values Count:\n", missing_counts)
print("\nProportion of Missing Values:\n", missing_proportion)
```
