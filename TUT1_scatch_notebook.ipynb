```python
import pandas as pd

# URL for the Wine Quality dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

# Load the dataset with proper delimiter
wine_df = pd.read_csv(url, delimiter=';')

# Check for missing values
missing_values = wine_df.isnull().sum()
print("Missing values in each column:")
print(missing_values)

# Check for any missing data in the entire DataFrame
total_missing = wine_df.isnull().sum().sum()
print(f"\nTotal missing values in the dataset: {total_missing}")

# Optional: Check percentage of missing values in each column
missing_percentage = wine_df.isnull().mean() * 100
print("\nPercentage of missing values in each column:")
print(missing_percentage)
```
